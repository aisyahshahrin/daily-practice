```python
set_one={1,2,3,4}
set_two={5,6,4}
```

```python
print('Number of Item in set_one',len(set_one))
print("Number of item in set_two", len(set_two))
common=set_one.intersection(set_two)
print('common items in both sets :',common)
```

```python

#gabungan 2 set
all_item=set_one.union(set_two)
print('all items in both sets :',all_item)
```

```python
diff_item=set_one.difference(set_two)
print('Items in set_one but not in set_two:', diff_item)
```

```python
course_one={'Python','Power BI','Excel','SQL','Time Management','Soft Skill'}
```

```python
course_two={'Python','Excel','SQL','Skill','Azure','GitHub','Tensorflow'}
```

```python
print('number of items in course_one', len(course_one))
for item in course_one:
  print(item)
```

```python
print('number of items in course_two:-', len(course_two))
for item in course_two:
  print(item)
```

```python
common_topics=course_one.intersection(course_two)
print('number of common topics in both courses:', len(common_topics))
print('common topics in both courses:', common_topics)
```

```python
both_courses=course_one.union(course_two)
print('number of topics in both courses:', len(both_courses))
print('topics in both courses:',both_courses)
```

```python
course_one_diff=course_one.difference(course_two)
print('topics in course_one but not course_two:',len(course_one_diff))
print('topics in course_one but not course_two:',course_one_diff)
```

```python
course_two_diff=course_two.difference(course_one)
print('topics in course_two but not course_one:',len(course_two_diff))
print('topics in course_two but not course_one:',course_two_diff)
```
