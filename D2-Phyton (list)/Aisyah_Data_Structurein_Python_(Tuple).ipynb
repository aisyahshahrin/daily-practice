Tupe: is and ordered Immutable collection of Items


*   We use () to denote tuple



```python
numbers=(100,130,100,500,600)
```

```python
print('Numbers of Item In tuple are:',len(numbers))
```

```python
print('All Items in tuple')
for item in numbers:
  print(item,end='\t')
```

```python
for (index,item) in enumerate(numbers):
  print(f'{index+1}-->{item}')
```

```python
for (index,item) in enumerate(numbers):
  print(f'{index}-->{item}')
```
