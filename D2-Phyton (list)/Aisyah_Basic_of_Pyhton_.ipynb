```python
#continue: skip that particular iteration adn loop will continue after that

for i in range(1,100):
  if (i%5==0):
   continue

print(i,end='\t')
```

```python
#break:it will breakdown the loop
for i in range(1,100):
 if (i%5==0):
   break

 print(i,end='\t')
```

```python
atm_pin=1234
allowed_attempts=3

while(allowed_attempts!=0):
  user_pin=int(input('enter your pin:\t'))
  if(user_pin==atm_pin):
    print('welcome to ATM')
    break

  else:
    allowed_attempts-=1
    if(allowed_attempts==0):
      print('your card has been blocked')
    else:
      print('wrong Pin Try Again')
```

```python
#cara 1
hidden_number=7
allowed_attempts=3

while(allowed_attempts!=0):
  user_pin=int(input('guess the number:\t'))
  if(user_pin==hidden_number):
    print('Correct answer You Have Won')
    break
  else:
    allowed_attempts-=1

  if(allowed_attempts==0):
      print('your have lost all your attempts')
  else:
    if(user_pin < hidden_number):
      print('Too Low')
    else: # user_pin > hidden_number
      print('Too High')
```
